```python
# not to do stemming

import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()
```

```python
def st(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
st(df_n["Tag"][0])
```

```python
df_n["Tag"]=df_n["Tag"].apply(st)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(df_n["Tag"]).toarray()
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
simulary=cosine_similarity(vectors)
```

# now if we give the index in the vector it give the simularities

```python
def simul(movie):
    index=df[df["title"]==movie].index[0]
    distance=simulary[index]
    #for 1st movie 2 movies simularities and 1st movie 3 movie simularities , 1 move 4th....
    list_v=sorted(list(enumerate(distance)),reverse=True,key=lambda x: x[1])[1:6]
    
    for i in list_v:
        c=i[0]
        print(c,df_n.iloc[c]["title"])
```
